## Ejercicio Extra — Frecuencia de palabras en un archivo de texto

Se pide analizar un archivo de texto para obtener la frecuencia de aparición de cada palabra.

1. Lee el contenido completo del archivo y extrae las palabras.
2. Normaliza el texto para que palabras equivalentes se contabilicen juntas: ignora las mayúsculas y elimina signos de puntuación.
3. Construye un diccionario donde cada clave sea una palabra y su valor el número de veces que aparece en el texto.
4. Muestra el listado de palabras ordenado de mayor a menor frecuencia.

````python
import string
from collections import Counter

def contar_palabras(archivo):
    """
    Lee un archivo de texto y cuenta la aparición de cada palabra.
    """
    try:
        with open(archivo, "r", encoding="utf-8") as f:
            texto = f.read().lower()

        # Eliminar puntuación
        texto = texto.translate(str.maketrans("", "", string.punctuation))

        # Dividir en palabras
        palabras = texto.split()

        # Contar ocurrencias
        frecuencia_palabras = Counter(palabras)
        return dict(frecuencia_palabras)

    except FileNotFoundError:
        print("Error: El archivo no existe.")
        return {}
    except Exception as e:
        print(f"Ocurrió un error: {e}")
        return {}

def imprimirDiccionario(frecuencia_palabras):
    # Mostrar resultados ordenados por frecuencia (mayor a menor)
    for palabra, frecuencia in sorted(frecuencia_palabras.items(),
                                      key=lambda x: x[1],
                                      reverse=True):
        print(f"{palabra}: {frecuencia}")

print("Inicio")
archivo = "texto_test.txt"
freq = contar_palabras(archivo)
imprimirDiccionario(freq)
````
