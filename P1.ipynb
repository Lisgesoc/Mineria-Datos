## Ejercicio 1 - Minutos de escucha por género musical

Tienes el historial de reproducciones de un usuario de Spotify. Cada registro contiene dos listas paralelas: una con los géneros musicales y otra con los minutos escuchados en cada sesión:

```python
generos = ["hiphop", "pop", "hiphop", "trap", "pop", "k-pop"]
minutos = [45, 30, 60, 25, 15, 40]
```

Tu objetivo es calcular el **tiempo medio de escucha por género**. Para ello:

1. Agrupa todas las sesiones correspondientes al mismo género.
2. Calcula la media de minutos asociados a cada uno.
3. Devuelve el resultado en forma de diccionario, donde la clave sea el género y el valor la media calculada.

Pista: Para saber si una clave existe en un diccionario se utiliza `in` o `not in` si consultamos que no exista.

````python
generos = ["hiphop", "pop", "hiphop", "trap", "pop", "k-pop"]
minutos = [45, 30, 60, 25, 15, 40]
````

### Solución

````python
#Solución ...

tam= len(generos)
dict = {}
dict2={}
for i in range(tam):
    if(dict.get(generos[i])):
        x = dict.get(generos[i])
        y =dict2.get(generos[i])
        x=(x*y+minutos[i])/(y+1)
        y+=1
        dict.update({generos[i]:x})
        dict2.update({generos[i]:y})
    else:
        dict[generos[i]]=minutos[i]
        dict2[generos[i]]=1

print(dict)
    
    
````

## Ejercicio 8 - Normalizar ratings de canciones

Tienes ratings de canciones del álbum de [**Hijos de la Ruina**](https://es.wikipedia.org/wiki/Natos_y_Waor) formado por:
- [Natos y Waor](https://es.wikipedia.org/wiki/Natos_y_Waor) - dúo de rap español
- [Recycled J](https://es.wikipedia.org/wiki/Recycled_J) - rapero de Madrid

Los usuarios puntúan del 1 al 10:

```python
ratings = [7.8, 5.2, 9.5, 6.3]
```

Realiza las siguientes operaciones:

1. Redondea cada valor al número entero más cercano.
2. Normaliza los valores al rango **[0, 1]** usando min-max:

$$
x' = \frac{x - \min}{\max - \min}
$$

Devuelve dos listas: una con los valores redondeados y otra con los normalizados.

````python
ratings = [7.8, 5.2, 9.5, 6.3]
````

### Solución

````python
#Solución
redondeos=[]
norm=[]
maximo = max(ratings)
minimo = min(ratings)
for r in ratings:
    if(r%1>=0.5):
        redondeos.append((r//1)+1)
    else:
        redondeos.append(r//1)

    x=(r-minimo)/(maximo-minimo)
    norm.append(x)

print (redondeos)
print (norm)
````

## Ejercicio 9 - Rellenar datos faltantes de seguidores

Tienes un registro de seguidores diarios de influencers en Instagram, pero algunos días faltan datos (`None`):

- [AuronPlay](https://www.instagram.com/auronplay/) - streamer y youtuber
- [María Pombo](https://www.instagram.com/mariapombo/) - influencer

```python
seguidores = [("@auronplay", 2800000), ("@auronplay", None), ("@auronplay", 2820000), ("@mariapombo", None), ("@mariapombo", 3100000)]
```

Reemplaza cada `None` por el **último valor válido de ese mismo influencer** (forward fill).
Si no existe un valor previo válido, deja `None`.


````python
seguidores = [("@auronplay", 2800000), ("@auronplay", None), ("@auronplay", 2820000), ("@mariapombo", None), ("@mariapombo", 3100000)]
````

### Solución

````python
#Solución
dicSeg={}
i=0
for s in seguidores:
    if(dicSeg.get(s[0])==None):
        dicSeg.update({s[0]: s[1]})
    elif(s[1]==None and dicSeg[s[0]]!=None):
        seguidores[i]=(s[0],dicSeg[s[0]])
    i+=1

print (seguidores)

````
